```python
# Group data by issue_month and calculate metrics

# Ensure issue_date is parsed correctly
df['issue_date'] = pd.to_datetime(df['issue_date'], errors='coerce')
# Extract month-year
df['issue_month'] = df['issue_date'].dt.to_period('M')
# Aggregate metrics
monthly_trends = df.groupby('issue_month').agg(
    Total_Loan_Applications=('member_id', 'count'),
    Total_Funded_Amount=('loan_amount', 'sum'),
    Total_Amount_Received=('total_payment', 'sum')
).reset_index()
# Convert back to timestamp for plotting
monthly_trends['issue_month'] = monthly_trends['issue_month'].dt.to_timestamp()
# Plot line chart
fig = px.line(
    monthly_trends,
    x='issue_month',
    y=['Total_Loan_Applications', 'Total_Funded_Amount', 'Total_Amount_Received'],
    labels={'issue_month': 'Issue Month', 'value': 'Metrics Values', 'variable': 'Metrics'},
    title="Monthly Trends by Issue Date"
)

fig.show()
```

```python
#Regional Analysis by State

# Aggregate metrics by state
state_trends = df.groupby('address_state').agg(
    Total_Loan_Applications=('member_id', 'count'),
    Total_Funded_Amount=('loan_amount', 'sum'),
    Total_Amount_Received=('total_payment', 'sum')
).reset_index()

# Filled map for Loan Applications
fig1 = px.choropleth(
    state_trends,
    locations='address_state',
    locationmode="USA-states",
    color='Total_Loan_Applications',
    scope="usa",
    title="Total Loan Applications by State"
)
fig1.show()

# Filled map for Funded Amount
fig2 = px.choropleth(
    state_trends,
    locations='address_state',
    locationmode="USA-states",
    color='Total_Funded_Amount',
    scope="usa",
    title="Total Funded Amount by State"
)
fig2.show()

# Filled map for Amount Received
fig3 = px.choropleth(
    state_trends,
    locations='address_state',
    locationmode="USA-states",
    color='Total_Amount_Received',
    scope="usa",
    title="Total Amount Received by State"
)
fig3.show()
```

```python
df['term']=df['term'].str.replace('months','')
df['verification_status']=df['verification_status'].str.replace('Source Verified','Verified')
```

```python
#Loan Term Analysis
# Aggregate metrics by loan term
term_trends = df.groupby('term').agg(
    Total_Loan_Applications=('member_id', 'count'),
    Total_Funded_Amount=('loan_amount', 'sum'),
    Total_Amount_Received=('total_payment', 'sum')
).reset_index()
# Donut chart for Loan Applications
fig1 = px.pie(
    term_trends,
    values='Total_Loan_Applications',
    names='term',
    hole=0.5,   # makes it a donut chart
    title="Loan Applications Distribution by Term"
)
fig1.show()
# Donut chart for Funded Amount
fig2 = px.pie(
    term_trends,
    values='Total_Funded_Amount',
    names='term',
    hole=0.5,
    title="Funded Amount Distribution by Term"
)
fig2.show()
# Donut chart for Amount Received
fig3 = px.pie(
    term_trends,
    values='Total_Amount_Received',
    names='term',
    hole=0.5,
    title="Amount Received Distribution by Term"
)
fig3.show()
```

```python
emp_length_trends = df.groupby('emp_length').agg(
    Total_Loan_Applications=('member_id', 'count'),
    Total_Funded_Amount=('loan_amount', 'sum'),
    Total_Amount_Received=('total_payment', 'sum')
).reset_index()
fig1 = px.bar(
    emp_length_trends,
    x='emp_length',
    y='Total_Loan_Applications',
    title="Total Loan Applications by Employee Length"
)
fig1.show()
fig2 = px.bar(
    emp_length_trends,
    x='emp_length',
    y='Total_Funded_Amount',
    title="Total Funded Amount by Employee length"
)
fig2.show()
fig3 = px.bar(
    emp_length_trends,
    x='emp_length',
    y='Total_Amount_Received',
    title="Total Amount Recieved by Employee Length"
)
fig3.show()
```

```python
#Loan Purpose Breakdown
purpose_trends = df.groupby('purpose').agg(
    Total_Loan_Applications=('member_id', 'count'),
    Total_Funded_Amount=('loan_amount', 'sum'),
    Total_Amount_Received=('total_payment', 'sum')
).reset_index()
purpose_trends_melted = purpose_trends.melt(
    id_vars='purpose',
    value_vars=['Total_Loan_Applications', 'Total_Funded_Amount', 'Total_Amount_Received'],
    var_name='Metric',
    value_name='Value'
)
fig = px.bar(
    purpose_trends_melted,
    x='purpose',
    y='Value',
    color='Metric',
    barmode='group',
    labels={'purpose': 'Loan Purpose', 'Value': 'Metrics Values', 'Metric': 'Metrics'},
    title="Loan Purpose Breakdown"
)
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
#Home Ownership Analysis
home_trends = df.groupby('home_ownership').agg(
    Total_Loan_Applications=('member_id', 'count'),
    Total_Funded_Amount=('loan_amount', 'sum'),
    Total_Amount_Received=('total_payment', 'sum')
).reset_index()
fig1 = px.treemap(
    home_trends,
    path=['home_ownership'],
    values='Total_Loan_Applications',
    title="Home Ownership Analysis - Loan Applications",
    color='Total_Loan_Applications',
    color_continuous_scale='blues'
)
fig1.show()
# TreeMap for Funded Amount
fig2 = px.treemap(
    home_trends,
    path=['home_ownership'],
    values='Total_Funded_Amount',
    title="Home Ownership Analysis - Funded Amount",
    color='Total_Funded_Amount',
    color_continuous_scale='Greens'
)
fig2.show()
#Total Amount Recieved
fig3 = px.treemap(
    home_trends,
    path=['home_ownership'],
    values='Total_Amount_Received',
    title="Home Ownership Analysis - Amount Received",
    color='Total_Amount_Received',
    color_continuous_scale='Purples'
)
fig3.show()
```
